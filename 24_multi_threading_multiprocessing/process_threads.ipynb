```python
#Practical implementation oof multithreading 
"""Whhen to use multithreading 

IO bound  task 
Conncurrent execution 

"""

import time

def print_numbers():
    for i in range(5):
        print(f"Number: {i}")
        time.sleep(1)

def print_letters():
    for letter in "abcde":
        print(f"Letter: {letter}")
        time.sleep(1)

start_time = time.time()

print_numbers()
print_letters()

finished_time = time.time() - start_time
print(f"Time taken (without multithreading): {finished_time}")
```

```python
##Execution by creating thereading 

import threading
import time

def print_numbers():
    for i in range(5):
        print(f"Number: {i}")
        time.sleep(1)

def print_letters():
    for letter in "abcde":
        print(f"Letter: {letter}")
        time.sleep(1)

start_time = time.time()

# Create threads
t1 = threading.Thread(target=print_numbers)
t2 = threading.Thread(target=print_letters)

# Start threads
t1.start()
t2.start()

# Wait for threads to finish(complete execution )
t1.join()
t2.join()

finished_time = time.time() - start_time
print(f"Time taken (with multithreading): {finished_time}")
```
