What is an Outlier?

An outlier is a data point that is significantly different from the rest of the data.
It lies far away from the majority of observations and can distort analysis and model performance.

Example:

Dataset (Salary in ₹):

25,000  30,000  28,000  27,000  1,000,000


Why Are Outliers a Problem?

Outliers can:

Skew mean and standard deviation

Mislead machine learning models

Increase error in regression

Affect distance-based algorithms (KNN, K-Means)

Reduce model accuracy

Causes of Outliers

Data entry errors (extra zero, wrong unit)

Measurement errors

Natural variation (real extreme events)

Sampling issues

Fraud or anomaly behavior

Types of Outliers
1️⃣ Global Outliers

Extremely different from entire dataset
Example: Salary = 1,000,000

2️⃣ Contextual Outliers

Abnormal only in a specific context
Example: High temperature at night

3️⃣ Collective Outliers

Group of unusual data points
Example: Sudden spike in network traffic

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    "salary": [25000, 28000, 30000, 32000, 35000, 40000, 45000, 50000, 1000000]
}

df = pd.DataFrame(data)
df
```

```python
plt.boxplot(df["salary"])
```

```python
plt.title("Salary Boxplot ")

plt.show()
```

```python
#IQR method for outlier detect
Q1=df["salary"].quantile(0.25)
Q3=df["salary"].quantile(0.75)

IQR=Q3-Q1

lower_limit=Q1 -1.5*IQR
upper_limit=Q3 + 1.5*IQR

Q1,Q3,IQR,lower_limit,upper_limit
```

```python
#Outliers identificaton 
outliers = df[(df["salary"] < lower_limit) | (df["salary"] > upper_limit)]
outliers
```

```python
df_no_outliers = df[(df["salary"] >= lower_limit) & (df["salary"] <= upper_limit)]
df_no_outliers
```

```python
print("Mean with outlier:", df["salary"].mean())
print("Mean without outlier:", df_no_outliers["salary"].mean())
```
