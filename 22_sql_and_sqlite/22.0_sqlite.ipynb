SQL and SQLite 

SQL (Structured Query Language ) is standard language of managigng and manipulating relational databases. SQlite is a self contained , serverless and zero configuration database engine that is widely used for embeddded datbase systems. In this lesson , we will cover the basics of sql and sqlite , inclding crating databases, tables and performing various sql operations. 
 


```python
import sqlite3
```

```python
##connect to an SQLite database 
connection=sqlite3.connect('example.db')
connection
```

```python
cursor=connection.cursor()
```

```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS employees (
    emp_id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    age INTEGER,
    department TEXT,
    salary REAL
)
''')

##commmmit the changes
connection.commit()
```

```python
cursor.execute(
    '''
    select * from employees
    '''
)
```

```python
##insert data teh sqlite table 
cursor.execute('''
Insert Into employees(name,age,department)

values('Deepak',26,'Data science')

''')
cursor.execute('''
Insert Into employees(name,age,department)

values('smith',26,'healthcare')

''')
cursor.execute('''
Insert Into employees(name,age,department)

values('Sumit',32,'Finance')

''')

##connection 
connection.commit()
```

```python
##Query the data from the table 

cursor.execute('select*from employees')
```

```python
rows=cursor.fetchall()

##Print the queried data 
for row in rows:
    print(row)
```

```python
cursor.execute(
    '''
    UPDATE employees
    SET age=23
    where name="rajan" 
    '''
)
```

```python
##delete the data form teh table 

cursor.execute('''
Delete from employees where name='Deepak'


''')
```

```python
##working with sales data 
#connect to an sqllite database 
connection=sqlite3.connect('saledata.db')
cursor=connection.cursor()

#create a table for sales data
cursor.execute(
    '''
    CREATE TABLE IF NOT EXISTS sales (
    id INTEGER PRIMARY KEY,
    date TEXT NOT NULL,
    product TEXT NOT NULL,
    sales INTEGER,
    region TEXT)
    '''
)
```

```python
sales_data = [
    ('2025-01-16', 'Mobile', 12, 'South'),
    ('2025-01-17', 'Tablet', 6, 'East'),
    ('2025-01-18', 'Laptop', 4, 'West'),
    ('2025-01-19', 'Headphones', 20, 'North')
]

cursor.executemany(
    '''
    INSERT INTO sales (date, product, sales, region)
    VALUES (?, ?, ?, ?)
    ''',
    sales_data
)

connection.commit()
```
