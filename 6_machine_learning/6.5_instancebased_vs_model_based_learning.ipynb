üìå What is Instance-Based Learning?

Instance-based learning is a lazy learning approach where the algorithm stores the entire training data and makes predictions only when a new query (test point) arrives.

üëâ The model does not learn a general rule in advance.
üëâ It compares the new data point with stored instances.

2.How it Works (Step-by-Step)

Store all training examples in memory

When a new data point comes:

Measure similarity (distance) with stored data

Select nearest instances

Predict output based on them

Example

Suppose you want to predict house price:

Find houses similar in size, location

Predict price based on nearby houses

What is Model-Based Learning?

Model-based learning is an eager learning approach where the algorithm learns a mathematical model from training data and then uses that model to make predictions.

üëâ Data is summarized into a model (equation/tree/network)
üëâ Original data is not needed at prediction time

Exam-Ready Answer

Instance-based learning stores training instances and makes predictions by comparing new data with existing data using similarity measures. It is a lazy learning approach and requires high memory and computation during prediction.

Model-based learning, on the other hand, builds a generalized model during training and uses it for prediction. It is faster at prediction time, uses less memory, and generalizes better.

Examples of instance-based learning include KNN, while model-based learning includes regression and decision tree algorithms.
