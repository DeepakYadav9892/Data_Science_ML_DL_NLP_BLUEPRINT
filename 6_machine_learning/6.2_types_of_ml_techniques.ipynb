2. Unsupervised Learning
Definition

Unsupervised Learning is a type of machine learning where no labeled data is provided.
The model finds patterns, relationships, or structures in the data by itself.

Example

Customer data is given, but it is not known in advance which customers are premium or regular.

Types of Unsupervised Learning
(a) Clustering

Groups similar data points together

Examples:

Customer segmentation

Market research

Student grouping

Algorithms:

K-Means

Hierarchical Clustering

DBSCAN

(b) Association Rule Learning

Finds relationships between variables in large datasets

Example:

Customers who buy bread also buy butter

Algorithms:

Apriori

Eclat

Applications

Recommendation systems (Amazon, Netflix)

Market basket analysis

Customer behavior analysis

3. Semi-Supervised Learning
Definition

Semi-Supervised Learning uses both labeled and unlabeled data.
Usually, a small portion of data is labeled, while a large portion is unlabeled.

Why it is used?

Labeling data is expensive and time-consuming

Unlabeled data is easily available

Example

In an image dataset:

100 images are labeled

900 images are unlabeled

The model learns from labeled data first and then improves using unlabeled data.

Applications

Image recognition

Speech recognition

Medical diagnosis

4. Reinforcement Learning
Definition

Reinforcement Learning is a type of machine learning where an agent learns by interacting with an environment using trial and error.

The agent receives:

Reward for correct actions

Penalty for wrong actions

The goal is to maximize total reward.

Key Components

Agent

Environment

Action

Reward

Example

In a game:

Correct move ‚Üí +10 reward

Wrong move ‚Üí ‚àí5 reward

Applications

Policy Gradient Methods
