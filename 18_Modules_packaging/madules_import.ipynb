Importing modules in python : modules and packages 

In python, modules and packages help organize and resue code. Here's comprehencive guide on how to import them . 

```python
import math
math.sqrt(16)
```

```python
from math import sqrt,pi
print(sqrt(64))
print(sqrt(25))
print(pi)
```

```python
import numpy as np
np.array([1,2,3,4])
```

```python
from math  import * 
sqrt(16)
print(pi)
```

```python
from package.maths import addition
addition(2,3)
```

```python
from package import maths
maths.addition(89,45)
```
