```python
grouped_mean=df.groupby('Product Name')['Unit Price'].mean()
```

```python
print(grouped_mean)
```

```python
grouped_sum=df.groupby(['Product','Region'])['Value'].sum()
print(grouped_sum)
```

```python
df.groupby(['Product','Region'])['Value'].mean()
```

```python
##aggreagate multiple functions
grouped_agg=df.groupby('Region')['Value'].agg(['mean','sum','count'])
grouped_agg
```

```python
##mrrgingg and joining dataframes 
df1=pd.DataFrame({'Key':['A','B','C'],'Value1':[1,2,3]})

df2=pd.DataFrame({'Key':['A','B','D'],'Value2':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
##mearge dataframe onn the key columns 
pd.merge(df1,df2,on="Key",how="inner")
```

```python
pd.merge(df1,df2,on="Key",how="outer")
```

```python
pd.merge(df1,df2,on='Key',how="left")
```

```python
pd.merge(df1,df2,on='Key',how="right")
```

```python
pd.merge(df1,df2,on='Key',how="inner")
```
