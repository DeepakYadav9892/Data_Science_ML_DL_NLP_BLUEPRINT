```python
import matplotlib.pyplot as plt

```

```python
x=[1,2,3,4,5,6]
y=[10,20,30,42,45,60]
```

```python
plt.plot(x,y)

plt.xlabel('X axis')
plt.ylabel('Y axis ')

plt.title("Basic line plot ")
```

```python
x=[1,2,3,4,5,6]
y=[10,20,30,42,45,60]
##Create a customised line plot 
plt.plot(x,y,color='red',linestyle='-.',marker='*',linewidth=3,markersize=9)
plt.grid(True)
```

```python
##Histogram 
""" Histogram are used to represent the distribution of a dataset . they divide the data into bins 
and count the number of data points in each bin  """
##Bar plot 
marks = [45, 55, 60, 65, 70, 72, 75, 78, 80, 85, 90, 95]

import matplotlib.pyplot as plt

plt.hist(marks, bins=5,color='orange',edgecolor='black')
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Histogram of Marks")

plt.show()

```

```python
subjects = ["Maths", "Physics", "Chemistry", "Biology"]
students = [50, 40, 45, 35]


import matplotlib.pyplot as plt

plt.bar(subjects, students,color='purple')
plt.xlabel("Subjects")
plt.ylabel("Number of Students")
plt.title("Students in Each Subject")

plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,3,4,5,6]

plt.scatter(x,y,color='blue',marker='x')
```

```python
import matplotlib.pyplot as plt

subjects = ["Maths", "Physics", "Chemistry", "Biology"]
students = [40, 30, 20, 10]

colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.2, 0, 0, 0)

plt.pie(
    students,
    labels=subjects,
    colors=colors,
    explode=explode,
    autopct="%1.1f%%",
    shadow=True
)

plt.title("Students Distribution")
plt.axis("equal")   # perfect circle banane ke liye
plt.show()
```
