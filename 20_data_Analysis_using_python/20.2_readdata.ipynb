```python
import pandas as pd
from io import StringIO

Data = """
{
  "employee_name": "James",
  "email": "deepak@gmail.com",
  "job_profile": [
    {
      "title1": "TeamLead",
      "title2": "Sr.Developer"
    }
  ]
}
"""
df=pd.read_json(StringIO(Data))
```

```python
df.to_json()
```

```python
df.to_json(orient='index')
```

```python
df.to_json(orient='records')
```
