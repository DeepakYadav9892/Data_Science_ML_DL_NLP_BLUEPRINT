Inheritance in python 

Inheritance is a fundamental concept in Object oriented programming (OOP) that allows a class to inherit attributes and methods fromm another class. this lesson covers single ingeritance and mulltiple inheritance, demonstrating hwo to ccreate and use them in python. 


```python
###Inheritence 
##parent class 
class Car:
    def __init__(self,windows,door,enginetype):
        self.windows=windows
        self.door=door
        self.enginetype=enginetype

    def drive(self):
        print("the person will drive the {self.enginetype} car")
```

```python
car1=Car(4,5,'petrol')
print(car1.drive())
```

```python
class Tesla(Car):
    def __init__(self,windows,doors,enginetype,is_selfdriving):
        super().__init__(windows,doors,enginetype)
        self.is_selfdriving=is_selfdriving
    def selfdriving(self):
        print("Tesla supports {self.selfdriving}")
```

```python
tesla1=Tesla(4,5,"electric",True)
tesla1.selfdriving()
```

```python
tesla1.drive()
```

```python
##Multiple inheritence 
##whhen a clas inherits fromm more thean one base class . 
class Animal:
    def __init__(self,name):
        self.name=name
    def speak(self):
        print("Subclass must implement thiis method")

##Base class 2 
class Pet:
    def __init__(self,owner):
        self.owner=owner


##Derived clas 
class Dog(Animal,Pet):
    def __init__(self, name,owner):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)
    def speak(self):
        return f"{self.name} says woof "
    
dog=Dog("Buddy","deepak")
print(dog.speak())
print(dog.owner)
```
