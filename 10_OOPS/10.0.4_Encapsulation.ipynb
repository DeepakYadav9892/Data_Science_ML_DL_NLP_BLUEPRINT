Encapsulation And Abstraction 



Encapsulation and abstraction are two fundamental principles of object Oriented Programmiing that help is designing robust,maintanable,  and reusable code. Encapsulation involves bundling data and methods that operate on the data within a ssingle unit, while abstraction involves hiding complex implementstaion details and exposing only the necessary features. 



Encapsulation 

Encasulation is the concept of wrapping data (variables) and methods (functions ) together  as a single unit. It restricts direct access to some of the objects componets , which is means of preventing accidental interference and misuse of the data.

```python
class Person:
    def __init__(self,name,age,gender):
        self._name=name #protected variabale created by usign sigle underscore before the name of the varibale 

        
        self._age=age #protected  variables  
        self.gender=gender
class Employee(Person):  ##protected variable can be access from the derived class outside class without being derived its cant be access thhe protecte class 

    def __init__(self, name, age, gender):
        super().__init__(name, age, gender)
def get_name(person):
    return person._name

employee=Employee("Deepak",45,"Male")
print(employee._name)
```

```python
##Encapsualation with getter and setter methods 
class Person:
    def __init__(self,name,age):
        self.__name=name ##private access modifier or variable
        self.__age=age ##Private variable 

    ##getter method for name 
    def  get_name(self):
        return self.__name
    
    ##setter method yfor name 
    def set_name(self,name):
        self.__name=name
    def set_age(self,age):
        if age>0:
            self.__age=age
        else:
            print("Age canot be negative")
    
    def get_age(self):
        return self.__age
    
person=Person("Deepak",25)
print(person.get_name())
print(person.get_age())
person.set_age(35)
print(person.get_age())

person.set_age(-5)
        
```
