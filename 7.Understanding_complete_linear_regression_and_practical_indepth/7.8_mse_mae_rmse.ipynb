In regression problems, our model predicts continuous values (price, salary, temperature).
To check how good these predictions are, we use error metrics.

The three most important ones are:

MAE ‚Äì Mean Absolute Error

MSE ‚Äì Mean Squared Error

RMSE ‚Äì Root Mean Squared Error

1Ô∏è‚É£ Mean Absolute Error (MAE)
üìå Definition

MAE is the average of absolute differences between actual and predicted values.


Used when the output is **continuous** (price, salary, temperature).

---

### 1. Mean Absolute Error (MAE)

$$
\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} \left| y_i - \hat{y}_i \right|
$$

‚úî Easy to understand  
‚ùå Does not penalize large errors strongly  

---

### 2. Mean Squared Error (MSE)

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} \left( y_i - \hat{y}_i \right)^2
$$

‚úî Penalizes large errors  
‚ùå Units are squared (less intuitive)  

---

### 3. Root Mean Squared Error (RMSE)

$$
\text{RMSE} = \sqrt{\text{MSE}}
$$

‚úî Same units as target variable  
‚úî Most popular regression metric  

---

### 4. R¬≤ Score (Coefficient of Determination)

$$
R^2 = 1 - \frac{SS_{res}}{SS_{tot}}
$$

**Meaning:**
- \( R^2 = 1 \) ‚Üí Perfect model  
- \( R^2 = 0 \) ‚Üí No improvement over mean  

‚úî Explains how much variance is captured by the model
