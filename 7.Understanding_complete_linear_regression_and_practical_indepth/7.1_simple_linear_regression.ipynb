What is Linear Regression?

Linear Regression is a supervised learning algorithm used to model the relationship between:

Independent variable (X) and

Dependent variable (y)

It assumes this relationship is linear.

Simple Linear Regression

When there is one independent variable, it’s called Simple Linear Regression.

Mathematical Model
y=mx+c
Where:

y = dependent variable (output)

x = independent variable (input)

m = slope (how much y changes when x increases by 1)

c = intercept (value of y when x = 0)

Intuition (Real-Life Example)

House Price Prediction

X = size of house (sq ft)

y = price

If m = 5000, then:

Every 1 sq ft increase → price increases by ₹5000

Assumptions of Linear Regression (VERY IMPORTANT)

Linearity: Relationship between X and y is linear

Independence: Observations are independent

Homoscedasticity: Constant variance of errors

Normality of Errors: Errors are normally distributed

No Multicollinearity (not relevant for simple LR)

How Does the Model Learn? (Training)

The model learns m and c by minimizing the error between actual and predicted values.

Error (Residual)

Residual=Y(actual)-Y(predicted)

Cost Function (Loss Function)

We use Mean Squared Error (MSE):

$$
J(m, c) = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$


### Optimization: Gradient Descent

Gradient Descent updates parameters step-by-step:

$$
m := m - \alpha \frac{\partial J}{\partial m}
$$

$$
c := c - \alpha \frac{\partial J}{\partial c}
$$


$$
m := m - \alpha \frac{\partial J}{\partial m}, \quad
c := c - \alpha \frac{\partial J}{\partial c}
$$


```python
# Practical Example (Hands-on)
#1 Create a Simple Dataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Simple dataset
X = np.array([1, 2, 3, 4, 5])
y = np.array([3, 5, 7, 9, 11])  # y = 2x + 1
```

```python
# Visualize the Data
plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
# 3 Train Simple Linear Regression

from sklearn.linear_model import LinearRegression

X = X.reshape(-1, 1)  # 2D array required
model = LinearRegression()
model.fit(X, y)

m = model.coef_[0]
c = model.intercept_

print("Slope (m):", m)
print("Intercept (c):", c)
```

```python
# 4: Make Predictions
y_pred = model.predict(X)
```

```python
# 5: Plot Regression Line
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.show()
```

```python
# Model Evaluation

# 1️⃣ Mean Squared Error
# from sklearn.metrics import mean_squared_error

from sklearn.metrics import mean_squared_error
mean_squared_error(y, y_pred)
```

```python

#2️⃣ R² Score (Goodness of Fit) 
from sklearn.metrics import r2_score
r2_score(y, y_pred)


# R² = 1 → perfect fit

# R² = 0 → no relationship
```

The model is trained by minimizing a cost function such as Mean Squared Error using optimization techniques like gradient descent.
It is widely used for prediction and trend analysis in real-world problems.
