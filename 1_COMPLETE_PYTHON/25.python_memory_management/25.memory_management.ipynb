5.Memory management Best pracitcwws 


6.use local variables:Local variables have a shorter lifespan and are freed sooner than global variables. 

7.Avaoid circular references : circular references can lead to memory leaks if not properly managed.

8.use generators: Generator produce items one at a time and only keep one item in memory at a time,
making them memory efficent. 

9.Expllcitly delte objects:Use thedel statement to delete variables and objects explicitly. 

10.profile memory usage. use memory   profiling tools like tracemalloc and memory_profiler to identify memory leaks and optimize memory usage 

```python
import gc
class Myobject:
    def __init__(self,name):
        self.name=name
        print(f"object {self.name} created ")

    def __del__(self):
        print(f"Object {self.name} deleted ")


##create a circualar references 

obj1=Myobject("obj1")
obj2=Myobject("obj2")

obj1.ref=obj2
obj2.ref=obj1


del obj1
del obj2

##Menually trigger the garbage collectionn 

gc.collect()
        
```

```python
##Generator for memory efficiency 
##generator alllows you to produce items one at a time , using memory efficently by only keepint on items
def generate_numbers(n):
    for i in range(n):
        yield i

#Using the generator 
for num in generate_numbers(100000):
    print(num)
    if num>10:
        break
```

```python
##profiling memory usage with tracemalloc 
import tracemalloc 
def create_list():
    return [i for i in range (1000)]

def main():
    tracemalloc.start()

    create_list()
    snapshot=tracemalloc.take_snapshot()
    top_stats=snapshot.statistics('lineno')

    print("[Top 10 ]")

    for stat in top_stats[:10]:
        print(stat)
```

```python
main()
```
