Python Logging 

Logging is crusial aspect of anyapplication , providign a way to track events ,errors and operational inforamtionn.Python's built in logging module offers a flexible framework for emitting log messages fro Python programms. In this llesson , we will cover the basics of logging , including how to configure logging ,log levels and bes practicses for using loggiging in python applicattions.

```python
import logging
##configure teh basic loggingg settings 

logging.basicConfig(level=logging.DEBUG)

##log messages 
logging.debug("This is a debug message")
logging.info("this is an info message")
logging.warning("this is a warning message")
logging.error("this is an error messages")
logging.critical("This a critical messages ")

```

Log levels :

Python's logging module has several log levels indicating the severity of events.The default levels are:

DEBUG:Details information , typically of interest only when diagnossing problems .
INFO: Confirmation that things are working as expected 
WARNING:An indication that something unexpected happened or indictative of some problem in the near future(eg disk space low ).the sofware is stilll working as expected. 
ERROR:Due to a more serious problem, the softeare ahs not been able to perform 
CRITICAL: A very serious error , indication that the program itself may be unable to continue running . 


```python
##cofiguring logging 
logging.basicConfig(
    filename='app.log',
    filemode='w',
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)

##log messages 
logging.debug("This is a debug message")
logging.info("this is an info message")
logging.warning("this is a warning message")
logging.error("this is an error messages")
logging.critical("This a critical messages ")
```
