Data Visualization with seaborn 

Seaborn is a python visualization libray based on matplotlib that provides a high level interface for drawing attractive and informative statistical graphics . Seaborn helps in creating complex visualizations with just a few line of code. In this lesson , we will cover the basics of seaborn , including creating various type of plots and customizing them.

```python
import seaborn as sns 
```

```python
##Basic Plotting with seaborn 

tips=sns.load_dataset('tips')
tips
```

```python
##create a scatter plot 
import matplotlib.pyplot as plt 
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scattter plot of Total Bill vs Tip ")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.lineplot(x='size', y='total_bill', data=tips)
plt.title("Line plot of total bill by size")

plt.show()
```

```python
##categorical variable 
##Bar plot 

sns.barplot(x='day',y='total_bill',data=tips)
plt.title("Bar plot of total bill by day ")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxenplot(x="day", y="total_bill", data=tips)
plt.title("Boxen Plot of Total Bill by Day")

plt.show()

"""20.0_numpy_intro.ipynb

Box Plot

Data ka distribution dikhata hai

Median, Quartiles, Outliers show karta hai

ðŸ”¹ Boxen Plot (Enhanced Box Plot)

Large datasets ke liye better

Zyada quantiles dikhata hai

Distribution ko zyada detail me show karta hai

ðŸ‘‰ sns.boxenplot() = advanced box plot

"""
```

```python
##Violin Plot 
##Data ka shape dikhata hai
# Bataata hai:
# ðŸ‘‰ data kahaan zyada concentrated hai
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("Violin Plot of Total Bill by Day")

plt.show()
```

```python
##Histograms 
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
##KDE plot 
sns.kdeplot(tips['total_bill'],shade=True)
```

```python
#Pairplot 
sns.pairplot(tips)
```

```python
##Heat map 

tips
```

```python
##Heat Map 

corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
