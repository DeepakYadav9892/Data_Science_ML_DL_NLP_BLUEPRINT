Standard Library overview 

Python's standard library is a vast collection of modules and packages that come bundled with python , providing a wide range of functionalities out of the box . here's an overview of some of the most commenly used modules and packages standard library . 

```python
import array
arr=array.array('i',[1,2,3,4])
print(arr)
```

```python
import math
print(math.sqrt(64))
print(math.pi)
```

```python
#random 
import random 
print(random.randint(1,10))
print(random.choice(['apple','banana','cherry']))
```

```python
###Data serialization
import json
data={'name':'deepak','age':25}
json_str=json.dumps(data)
print(json_str)

print(type(json_str))
parsed_data=json.loads(json_str)
print(parsed_data)
print(type(parsed_data))
```

```python
##Csv 
import csv
with open('example.csv',mode='w',newline='') as file:
    writer=csv.writer(file)
    writer.writerow(['name','age'])
    writer.writerow(['deepak',26])

with open('example.csv',mode='r') as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)

```

```python
##datetime 
from datetime import  datetime,timedelta
now=datetime.now()
print(now)

yesterday=now-timedelta(days=1)
print(yesterday)
```

```python
##time 
import time 
print(time.time())
time.sleep(3)
print(time.time())
```

```python
##regulaar expression 
import re
pattern=r'\d+'
text='There are 123 apples 456'
match=re.search(pattern,text)
print(match.group())
```
