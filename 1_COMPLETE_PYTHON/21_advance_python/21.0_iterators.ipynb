Iterators 

Iterators are advanced pythonn concepts that allows for efficent looping and memory management .iterators provide a way to access elements of a collection sequentially without exposing the underlying structure. 


```python
my_list=[10,2,3,4,5,6]
for i in my_list:
    print(i)
```

```python
type(my_list)
```

```python
print(my_list)
```

```python
##Iteraor 
iterator=iter(my_list)
print(type(iterator))
```

```python
next(iterator)
next(iterator)
```

```python
iterator=iter(my_list)
try:
    next(iterator)
except StopIteration:
    print("there are no elements in the iterator")
```

```python
#String iterator 
my_string="Hello"
string_iterator=iter(my_string)
print(next(string_iterator))
print(next(string_iterator))
```
