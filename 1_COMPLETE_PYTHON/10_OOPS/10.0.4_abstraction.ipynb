Abstraction:

Abstraction is the concept of hiding the complex implemenntation details and showing 
onluu the necessary features of an object. This helps in reducing programming complexity  and efffort . 


```python
from abc import ABC,abstractmethod
##Abstract classes 

class Vehicle(ABC):
    def drive(self):
        print("The  vehhicle is used for driving ")

    @abstractmethod
    def start_engine(self):
        pass
class Car(Vehicle):
    def start_engine(self):
        print("Car enginne started")

    def operate_vehicle(vehicle):
        vehicle.start_engine()

car=Car()
operate_vehicle(car)
```
