```python
##Decorator 
def main_welcome(func):
  
    def sub_welcome_method():
        print("welcome to the advance python course")
        func()
    
        print("Please learn these conncept properly ")
    return sub_welcome_method()
```

```python
def course_introduction():
    print("this is an advance pythonn course ")
course_introduction()
```

```python
main_welcome(course_introduction)
```

```python
@main_welcome
def course_introduction():
    print("this is an advanced pyython course ")
```

```python
#Decorator 
def my_decorator(func):
    def wrapper():
        print("something is happening beffore the functionn is called")
        func()
        print("Something is happnng after the function is called ")
    return wrapper 
```

```python
@my_decorator
def say_hello():
    print("Hello!")
    
```

```python
##decorator with argumets 

def repeat(n):
    def decorator(func):
        def wrapper(*args,**kwargs):
            for _ in range(n):
                func(*args,**kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def say_hello():
    print("Hello")
```

```python
say_hello()
```
