Generators are a simple way to create iterrators. they use the yield keyword to produce  a series of vlaes lazily , which means they generate values on the fly and do not the flyy and do not store them in memory .

```python
def square(n):
    for i in range(3):
        yield i**2

print(square(10))
```

```python
for i in square(3):
    print(i)
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
    
```

```python
gen=my_generator()
gen
```

```python
next(gen)
```

```python
for val in gen:
    print(val)
```

##Practical :: Reading large files 

Generators are particularly useful for reading large files because thhey allow you to process one line at a time 
without loading the entire file into memory. 



```python
#practical reading large file s

def read_large_file(file_path):
    with open(file_path,'r')as file:
        for line in file:
            yield line
```

```python
file_path='large_file.txt'
for line in read_large_file(file_path):
    print(line.strip())
```
